## Using Matplotlib and Seaborn

### The end goal of Data visualization is,
* visual representation of data to discover trends and patterns
* helps in storytelling

### Types
* Categorical: Bar chart, Pie chart, etc
* Numerical: Box plot, Histogram,etc
* Relationships: Scatter plot, Line plot, Heat maps, Pair plot,etc

### Libraries
* Matplotlib
* Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### In Matplotlib, you need to understand,
* Figure: The entire canvas that holds all subplots. This can be related to an Excel workbook.
* Axes: A single plot. This can also be related to a single sheet in Excel workbook. Axes are found in figures just as worksheets are found in workbooks in Excel. This is where the x-axis and y-axis with the plottings are found.
* Subplot: When you plot a graph with two or more charts on a single sheet, the two charts combined are called subplots. EVERY SUBPLOT IS AN AXES, BUT NOT EVERY AXES IS A SUBPLOT. 

```python
fig, ax = plt.subplots()
ax.plot([1,2,3], [10,20,30])
ax.set_title('My First Axes')

plt.show()
```

```python
# Create a figure with a single axes

x = [1,2,3]

y = [4,5,6]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('My Basic Plot')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

plt.show() # To display the plot
```

```python
# Create a figure with multiple subplots

fig, axes = plt.subplots(1, 2) # This gives 2 plots side by side
axes[0].plot(x, y)
axes[1].plot(x, y)
axes[0].set_title('Left Plot')
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')

axes[1].set_title('Right Plot')
axes[1].set_xlabel('x-axis')
axes[1].set_ylabel('y-axis')

plt.show()
```

### Creating a plot with 4 axes

```python
fig, axes = plt.subplots(2, 2) # THis gives two rows and 2 columns

axes[0, 0].plot(x,y)
axes[0, 1].plot(x,y)
axes[1, 0].plot(x,y)
axes[1, 1].plot(x,y)

plt.show()
```

### What are the number of Netflix titles released per year?

```python
title_per_year = df.groupby('release_year')['title'].count().reset_index()

title_per_year
```

```python
# Since it is a numerical values,we use line graph

plt.plot(title_per_year['release_year'], title_per_year['title'])
plt.title('Number of Netflix Titles Released Per Year')
plt.xlabel('Released Year')
plt.ylabel('Number of Titles')

plt.show()
```

```python
# Using the fig and axes syntax to do the same thing

fig, axes = plt.subplots()
axes.plot(title_per_year['release_year'], title_per_year['title'], color='orange', marker='o')
axes.set_title('Number of Netflix Titles Released Per Year')
axes.set_xlabel('Released Year')
axes.set_ylabel('Number of Titles')

plt.show()
```

```python
type_categories = df['type'].value_counts()

type_categories
```

```python
# This is a categorical data so we cannot use line plot so we will use pie chart since we have only 2 categories.
# Only use bar chart when the categories are more than 3 to avoid it being clumsy on the pie chart
# Always use different shades of a color when drawing the charts

type_categories.index
```

```python
plt.pie(type_categories, labels=type_categories.index, autopct='%1.2f%%', colors=['navy', 'lightblue'])
plt.title('Distribution of Movies and TV Shows', color='blue')
plt.show()
```

## Bi-Variate Analysis
Examining the relationship between two variables in a dataset

```python
df['type'].value_counts()
```

```python
# Movies and Tv shows released per year
# The size method will count movies and shows differently for each year
# The unstack method will put the data series into data frame and fill the NaN values with 0
# The reset index method will include indexes instead of using release year as index

release_year_by_type = df.groupby(['release_year', 'type']).size().unstack(fill_value=0).reset_index()
release_year_by_type
```

```python
plt.plot(release_year_by_type['release_year'], release_year_by_type['Movie'], label='Movies')
plt.plot(release_year_by_type['release_year'], release_year_by_type['TV Show'], label='TV Shows')
plt.legend()

plt.show()
```
